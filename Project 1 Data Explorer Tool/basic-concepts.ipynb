```python
students = []

def add_student():
    name = input("Enter name: ")
    age = int(input("Enter age: "))
    marks = int(input("Enter marks: "))
    students.append({"name":name,"age":age,"marks":marks})

def has_no_students():
    if students:
        return False
    print("No students add students.")
    return True

def get_average():
    if has_no_students():
        return
    
    average = 0
    for student in students:
        average+=student["marks"]
    print(average/len(students))


def get_topper():
        if has_no_students():
            return
            
        print(max(students, key=lambda student: student["marks"]))

def print_all_students():
    if has_no_students():
        return
         
    for student in students:
        print(student)

def search_name():
    if has_no_students():
        return
                 
    name = input("Enter the name to search: ")

    for student in students:
        if(name == student["name"]):
            print(student)
            return
    
    print("No such student")

while True:
    print(
        """
===== Student Record Manager =====
1. Add Student
2. View All Students
3. Find Topper
4. Calculate Average Marks
5. Search Student by Name
6. Exit"""
    )
    option = int(input("Select option number from menu: "))
    match option:
        case 1:
            add_student()
        case 2:
            print_all_students()
        case 3:
            get_topper()
        case 4:
            get_average()
        case 5:
            search_name()
        case 6:
            break
        case _:
            print("No such option")
```

```python
# The -q flag stands for 'quiet'
%pip install pandas -q

import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "Age": [24, 30, 28, 35, 26],
    "Department": ["IT", "HR", "IT", "Sales", "HR"],
    "Salary": [50000, 45000, 60000, 55000, 48000]
}

df = pd.DataFrame(data)

print("****Task 1*****")
print("First 3 rows")
print(df.iloc[1:4])
print("Last 2 rows")
print(df.iloc[-2:])
print(f"Shape of the dataframe: {df.shape}")
print("Columns")
print(list(df.columns))
print("Info")
print(df.info())
print("Describe")
print(df.describe())


print("*******************Task 2***************")
print(df["Name"])
print(df["Age"])
print(df[["Name","Salary"]])

print("************Task 3************")
print(df.iloc[1])
print(df.iloc[-1])
print(df.iloc[1:3])
print(df.iloc[1:5,1:4])


print("************Task 4************")
print(df[df["Age"]>28])
print(df[df["Salary"]>50000])
print(df.query("Department == 'IT'"))

print("*************Task 5************")
df["Bonus"] = df["Salary"]*0.10
print(df)

print("**************Task 6************")
print(f"Average salary is: {df["Salary"].mean()}")
print(f"Highest salary is: {df["Salary"].max()}")
print(f"Lowest salary is: {df["Salary"].min()}")

print("***********Task 7***************")
print(df.groupby("Department"))
print(df.groupby("Department")["Name"].count())
```

```python
# The -q flag stands for 'quiet'
%pip install matplotlib -q

import matplotlib.pyplot as plt

print("******************Task 1 - Department wise employee count********************")
data = df.groupby("Department")["Name"].count()
plt.bar(data.index,data.values)
plt.xlabel("Department")
plt.ylabel("Count")
plt.grid(True)
plt.show()

print("******************Task 2 - Department distribution (Pie chart)********************")
plt.pie(data.values,labels=data.index,autopct = '%1.1f%%')
plt.show()


print("******************Task 3 - Salary distribution (Histogram)********************")
plt.hist(df["Salary"],bins = 3,color = "skyblue", edgecolor = "black")
plt.xlabel("Salary")
plt.ylabel("No. of employees")
plt.show()


print("******************Task 4 - Age vs Salary (Scatter Plot) ****************************")
plt.scatter(df["Age"],df["Salary"],color = "blue",marker = "o")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.grid(True)
plt.show()

print("******************Task 5 - Employee Salary (Line Chart) ****************************")
sorted_df = df.sort_values("Age")
plt.plot(sorted_df["Age"],sorted_df["Salary"],label = "Salary trend", marker = "o",color = "skyblue",linestyle = "-")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.grid(True)
plt.legend()
plt.show()

print("******************Task 5 - Bonus (Bar graph)****************************")
data_grouped = df.groupby("Department")["Salary"].mean()
plt.bar(data_grouped.index,data_grouped.values)
plt.xlabel("Department")
plt.ylabel("Salary")
plt.show()
```
