```python
import pandas as pd

df = pd.read_csv('/kaggle/input/datasets/dg00949/housing-dataset/Housing.csv')
```

```python
import numpy as np

print(f"Describe\n {df.describe()}")
print(f"\nColumns: {list(df.columns)}")
print(df.head())

X = df[["area",
    "bedrooms",
    "bathrooms",
    "stories",
    "parking"]].to_numpy()
Y = df["price"].to_numpy()
```

```python
#shuffling the data set
indices = np.random.permutation(len(X))

X = X[indices]
Y = Y[indices]
```

```python
#splitting
split_index = int(0.80 * len(X))

train_X = X[:split_index]
train_Y = Y[:split_index]
test_X = X[split_index:]
test_Y = Y[split_index:]
```

```python
#Feature Scaling
 
mean = np.mean(train_X,axis = 0)
std = np.std(train_X,axis = 0)
X_scaled = (train_X-mean)/std
```

```python
print(X_scaled.shape)
w = np.zeros(X_scaled.shape[1])
b = 0
epochs = 1000
learning_rate = 0.01

#model 
def predict(X,w,b):
    return X @ w + b

def loss(errors):
    # n = y_predicted.shape[0]
    return np.mean((errors)**2)

def dL_dw(errors,X):
    n = len(errors)
    return (2/n) * (X.T @ errors)

def dL_db(errors):
    return 2 * (np.mean(errors))

losses = []

for i in range(epochs):
    predicted = predict(X_scaled,w,b)
    errors = predicted - train_Y
    current_loss = loss(errors)
    gradient_w = dL_dw(errors,X_scaled)
    gradient_b = dL_db(errors)
    w = w - learning_rate*gradient_w
    b = b - learning_rate*gradient_b
    if i%100 == 0 :
        losses.append(current_loss)

losses = np.array(losses)

print(losses)
```

```python
#R Square Mean
def r_square(X,Y):
    ss_res = np.sum((Y-predict(X,w,b))**2)
    ss_total = np.sum((Y-np.mean(Y))**2)
    
    return 1 - (ss_res/ss_total)

print(r_square(X_scaled,train_Y))
```

```python
#80% Train 20% Test
print("r_square on test data")
#scaling test data
X_test_scaled = (test_X - mean)/std
print(r_square(X_test_scaled,test_Y))
```

```python
#Generate graphs
import matplotlib.pyplot as plt

epoch_numbers = range(0, epochs, 100)
plt.plot(epoch_numbers,losses)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training Loss")
plt.grid(True)
plt.show()
```

```python
pred = predict(X_test_scaled, w, b)
plt.scatter(test_Y, pred)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted")
plt.show()
```
