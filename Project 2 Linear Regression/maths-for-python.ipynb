```python
#Calculas

def f(x):
    return x**3

def dF(x):
    h = 1e-5
    return (f(x+h)-f(x))/h

def right_dF(x):
    return 3 * (x**2)

print(f(2))
print(f(4))
print(f(10))

print(dF(2))
print(dF(4))
print(dF(10))
print(right_dF(2))
```

```python
#gradient descent

#model 
import numpy as np


actual = 10
X = 2

def model(w,b,X):
    return w*X + b

def loss(predicted, actual):
    return (predicted-actual)**2

def dl_dw(predicted,actual):
    return 2*(predicted-actual)*X

def dl_db(predicted,actual):
    return 2*(predicted-actual)

def gradient(predicted,actual):
    return np.array([dl_dw(predicted,actual),dl_db(predicted,actual)])

w = 1.0
b = 0.0

learning_rate = 0.1

for i in range(20):
    print(f"******************--Iteration: {i}--*************")
    predicted = model(w,b,X)
    print(f"Prediction: {predicted}")
    l = loss(predicted,actual)
    print(f"Loss: {l}")
    gradient_vec = gradient(predicted,actual)
    print(f"Gradient: {gradient_vec}")
    w = w - learning_rate * gradient_vec[0]
    print(f"Weight: {w}")
    b = b - learning_rate * gradient_vec[1]
    print(f"Bias: {b}")
    if l == 0:
        break


    
```
